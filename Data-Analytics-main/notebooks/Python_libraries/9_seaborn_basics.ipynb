# Seaborn
### What is Seaborn?
- Seaborn is a Python data visualization library built on Matplotlib.
- Works well with Pandas DataFrames.
- Prettier & more advanced than Matplotlib.

### Why Use Seaborn in Data Analytics?
- Easier syntax for visualizations.
- Built-in themes for professional styling.
- Better handling of categorical data (e.g., bar plots, violin plots).
- Statistical analysis support (e.g., KDE plots, Regression plots).

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.barplot(x=["A", "B", "C"], y=[10, 20, 15])
plt.title("Bar Plot Example")
plt.show()
```

```python
data = pd.DataFrame({"Category": ["A", "B", "A", "C", "B", "A", "C", "B"]})
sns.countplot(x="Category", data=data, palette="Set2")
plt.title("Count Plot Example")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips, palette="coolwarm")
plt.title("Box Plot Example")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("Violin Plot Example")
plt.show()
```

```python
sns.pairplot(tips, hue="sex", palette="husl")
plt.show()
```

```python
tip = sns.load_dataset("tips").select_dtypes(include=['number'])
corr = tip.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="smoker", size="size", palette="muted")
plt.title("Scatter Plot Example")
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot Example")
plt.show()
```
